```python
import os
import glob
import pandas as pd

# Define the base directory where participant data is stored
base_dir = '/home/skander/ds004920/'

# Placeholder for storing file paths
data_files = []

# Loop through each participant's directory
for subj_dir in glob.glob(os.path.join(base_dir, 'sub-*')):
    subj_id = os.path.basename(subj_dir)
    print(f"Processing {subj_id}...")
    
    # Step 1: Find the .tsv file
    tsv_path = glob.glob(os.path.join(subj_dir, 'func', '*_events.tsv'))
    
    # Step 2: Find the BOLD .nii.gz file
    bold_path = glob.glob(os.path.join(subj_dir, 'func', '*_bold.nii.gz'))
    
    # Step 3: Find the phasediff .nii.gz file for field map correction
    phasediff_path = glob.glob(os.path.join(subj_dir, 'fmap', '*_phasediff.nii.gz'))
    
    # Store the retrieved paths
    if tsv_path and bold_path and phasediff_path:
        data_files.append({
            'subject': subj_id,
            'tsv': tsv_path[0],
            'bold': bold_path[0],
            'phasediff': phasediff_path[0]
        })

# Convert to DataFrame for easier handling
df_files = pd.DataFrame(data_files)
print(df_files)

# You can now use df_files for further processing
```
